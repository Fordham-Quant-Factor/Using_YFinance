```python
portfolio = yf.download(ticker_list,start = "2019-1-1", end = "2022-1-1")["Adj Close"]
```

```python
portfolio.head()
```

```python
portfolio.diff()
```

```python
portfolio.pct_change()
```

```python
portfolio.pct_change().round(3).dropna()
```

```python
portfolio.pct_change().mean().plot(kind = "bar", figsize = (12,8))
```

```python
import numpy as np
```

```python
log_rets = np.log(portfolio/portfolio.shift(1))
```

```python
log_rets
```

```python
log_rets.round(2).dropna().head(10)
```

```python
np.exp(log_rets.cumsum()).head(10)
```

```python
log_rets.cumsum().apply(np.exp)
```

```python
log_rets.cumsum().apply(np.exp).plot(figsize = (12,8), grid = True);
```
